# [24.07.17] Rebuttal: Generate Random Medium Parameters

```python
import numpy as np
import os
import pickle
from datetime import datetime

curtime=datetime.now().strftime("%y%m%d_%H%M%S")
fname = f"random_init_{curtime}"
n=50
list_albedo = []
list_sigmat = []
for i in range(n):
    init_albedo = np.stack((np.random.uniform(0.75,0.85,1),
                            np.random.uniform(0.80,0.90,1),
                            np.random.uniform(0.85,0.95,1),))
    init_albedo2 = np.random.permutation(init_albedo)
    init_albedo2 = init_albedo2.transpose((1,0))
    
    init_sigmat = np.stack((np.random.uniform(30,60,1),
                            np.random.uniform(55,85,1),
                            np.random.uniform(70,100,1),))
    init_sigmat2 = np.random.permutation(init_sigmat)
    init_sigmat2 = init_sigmat2.transpose((1,0))

    list_albedo.append(init_albedo2)
    list_sigmat.append(init_sigmat2)
    
init_albedos = np.concatenate(list_albedo)
init_sigmats = np.concatenate(list_sigmat)

print(f"STATS: mean {init_albedos.mean(axis=0)}, std: {init_albedos.std(axis=0)}")
print(f"STATS: mean {init_sigmats.mean(axis=0)}, std: {init_sigmats.std(axis=0)}")

pkl_init = {
    "n" : n,
    "init_albedo" : init_albedos,
    "init_sigmat" : init_sigmats,
    "mean_albedo" : init_albedos.mean(axis=0),
    "mean_sigmat" : init_sigmats.mean(axis=0),
    "std_albedo" : init_albedos.std(axis=0),
    "std_sigmat" : init_sigmats.std(axis=0),
}

with open(f'{fname}.pickle','wb') as fw:
    pickle.dump(pkl_init, fw)

# np.save(f"./{fname}_albedo",init_albedos)
# np.save(f"./{fname}_sigmat",init_sigmats)
print("Saved init albedo and sigmaT")
```
